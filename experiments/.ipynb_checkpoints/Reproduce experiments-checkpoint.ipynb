<h3>Compute pairwise distances between patients using the features</h3>

Only wassertein ATM, distances between graphs if lymphocyte_density

```python
# Object to fill
all_feature_names = list(list(all_features.values())[0].keys())
patients_distances = {feature_name : {distance_name : numpy.zeros((len(all_patients_ids), len(all_patients_ids))) for distance_name in ["EMD"]} for feature_name in all_feature_names}

# We iterate over features of interest
for feature_name in all_feature_names :
    print(feature_name)
    
    # We compare all patients distributions over the same interval (for using the same loss matrix)
    #flattened_all_distances = flatten([features[feature_name] for features in all_features.values()])
    #min_x = min(flattened_all_distances)
    #max_x = max(flattened_all_distances)

    # Let's compute pairwise distances using that feature
    for i in range(len(all_patients_ids)) :
        patient_i = all_patients_ids[i]
        feature_i = all_features[patient_i][feature_name]
        for j in range(i + 1, len(all_patients_ids)) :
            patient_j = all_patients_ids[j]
            feature_j = all_features[patient_j][feature_name]
            
            # Use EMD distance
            if feature_name =='lymphocyte_density':
                k1=list(feature_i.keys())
                v1=list(feature_i.values())
                k2=list(feature_j.keys())
                v2=list(feature_j.values())
                
                patients_distances[feature_name]["EMD"][i, j] = wasserstein_between_histograms(k1,v1,k2,v2)
            else:
                patients_distances[feature_name]["EMD"][i, j] = wasserstein(feature_i,feature_j)
            patients_distances[feature_name]["EMD"][j, i] = patients_distances[feature_name]["EMD"][i, j]
            
            # Use KS distance
            #patients_distances[feature_name]["KS"][i, j] = kolmogorov_smirnov(feature_i, feature_j)
            #patients_distances[feature_name]["KS"][j, i] = patients_distances[feature_name]["KS"][i, j]
    
            # Use KbL divergence
            #patients_distances["KL"][i, j] = kullback_leibler(feature_i, feature_j, min_x=min_x, max_x=max_x)
            #patients_distances["KL"][j, i] = kullback_leibler(feature_j, feature_i, min_x=min_x, max_x=max_x)

            
patients_distances['distance_between_lymphocytes_blobs']['EMD']= numpy.where(numpy.isnan(patients_distances['distance_between_lymphocytes_blobs']['EMD']),numpy.inf,patients_distances['distance_between_lymphocytes_blobs']['EMD'])
```

<h3>Rescaling with a MinMaxScaler</h3>

```python
for feature in all_feature_names:
    print(feature)
    dist=patients_distances[feature]['EMD']
    patients_distances[feature]['EMD']=(dist - dist[dist != numpy.inf].min()) / (dist[dist != numpy.inf].max() - dist[dist != numpy.inf].min())
    
```

<h3>Cluster patients using the distances of unique features</h3>

```python
# We iterate over features of interest
dict_clusters={}
for feature_name in all_feature_names :
    
    # Let's cluster patients based on these distances
    for distance_name in patients_distances[feature_name] :

        # Header
        print("=" * 114)
        title = feature_name + " / " + distance_name + " / Hierarchical"
        print(" " * int((114 - len(title)) / 2) + title)
        print("=" * 114)

        # Using hierarchical clustering (merge clusters above threshold, find best threshold exhaustively)
        ratio_per_threshold = []
        p_value_per_threshold = []
        patient_clusters_per_threshold = []
        nb_clusters = 1
        while True :
            try :
                nb_clusters += 1
                patient_clusters = hierarchical_clustering(patients_distances[feature_name][distance_name], all_patients_ids, nb_clusters=nb_clusters)
                patient_clusters.sort(key=len)
                patient_clusters = [flatten(patient_clusters[:-1]), patient_clusters[-1]]
                p_value = log_rank_test(patient_clusters, clinical_data)[1][1][0]
                ratio = len(patient_clusters[-1]) / (len(patient_clusters[-1]) + len(patient_clusters[0]))
                ratio_per_threshold.append(ratio)
                p_value_per_threshold.append(p_value)
                patient_clusters_per_threshold.append(patient_clusters)

            except :
                break
        plot_curve(ratio_per_threshold, p_value_per_threshold, xlabel="Bottom cluster size", ylabel="Log-rank test", file_name=OUTPUT_DIR + "/clustering/" + feature_name + "/" + distance_name + "/hierarchical/thresholds.png")
        best_index = numpy.argmin(p_value_per_threshold)
        plot_survival_curves(patient_clusters_per_threshold[best_index], clinical_data, file_name=OUTPUT_DIR + "/clustering/" + feature_name + "/" + distance_name + "/hierarchical/survival.png")
        p_values = log_rank_test(patient_clusters_per_threshold[best_index], clinical_data)
        #plot_matrix(p_values, round_values=3, file_name=OUTPUT_DIR + "/clustering/" + feature_name + "/" + distance_name + "/hierarchical/log_ranks.png")
        print('p-value :',p_values[1][0][1])
        dict_clusters[feature_name]=patient_clusters_per_threshold[best_index]
        #pop_0 = len(patient_clusters_per_threshold[best_index][1])/len(all_patients_ids)
        #pop_1 = len(patient_clusters_per_threshold[best_index][0])/len(all_patients_ids)
        
        #dict_pvalues[str(feature_name)]:[p_values[1][0][1],pop_0,pop1]}


```

<h3>Compute centrals histograms and diverses statistics</h3>

```python
clusters_dists={}
for feature_name in all_feature_names :
    patients_cluster0_distances = {feature_name  : numpy.zeros((len(dict_clusters[feature_name][0]), len(dict_clusters[feature_name][0]))) for feature_name in all_feature_names}
    patients_cluster1_distances = {feature_name  : numpy.zeros((len(dict_clusters[feature_name][1]), len(dict_clusters[feature_name][1]))) for feature_name in all_feature_names}

    for i in range(len(patients_cluster0_distances[feature_name])):
        patient_i = dict_clusters[feature_name][0][i]
        feature_i = all_features[patient_i][feature_name]
        for j in range(i + 1, len(patients_cluster0_distances[feature_name])) :
            patient_j = dict_clusters[feature_name][0][j]
            feature_j = all_features[patient_j][feature_name]

            if feature_name =='lymphocyte_density':
                k1=list(feature_i.keys())
                v1=list(feature_i.values())
                k2=list(feature_j.keys())
                v2=list(feature_j.values())

                patients_cluster0_distances[feature_name][i, j] = wasserstein_between_histograms(k1,v1,k2,v2)
            else:
                patients_cluster0_distances[feature_name][i, j] = wasserstein(feature_i,feature_j)
            patients_cluster0_distances[feature_name][j, i] = patients_cluster0_distances[feature_name][i, j]
            
    for i in range(len(patients_cluster1_distances[feature_name])) :
        patient_i = dict_clusters[feature_name][1][i]
        feature_i = all_features[patient_i][feature_name]
        for j in range(i + 1, len(patients_cluster1_distances[feature_name])):
            patient_j = dict_clusters[feature_name][1][j]
            feature_j = all_features[patient_j][feature_name]

            if feature_name =='lymphocyte_density':
                k1=list(feature_i.keys())
                v1=list(feature_i.values())
                k2=list(feature_j.keys())
                v2=list(feature_j.values())

                patients_cluster1_distances[feature_name][i, j] = wasserstein_between_histograms(k1,v1,k2,v2)
            else:
                patients_cluster1_distances[feature_name][i, j] = wasserstein(feature_i,feature_j)
            patients_cluster1_distances[feature_name][j, i] = patients_cluster1_distances[feature_name][i, j]
    clusters_dists[feature_name]={}
    clusters_dists[feature_name]['0']=patients_cluster0_distances[feature_name]
    clusters_dists[feature_name]['1']=patients_cluster1_distances[feature_name]
    
min_dists_per_feature={}

for feature_name in all_feature_names:
    min_dists_per_feature[feature_name]={}
    cluster0=clusters_dists[feature_name]['0']
    cluster1=clusters_dists[feature_name]['1']
    max_value0 = cluster0[numpy.isfinite(cluster0)].max()+1
    max_value1 = cluster1[numpy.isfinite(cluster1)].max()+1
    for i in range(len(cluster0)):
        for j in range(len(cluster0)):
            if cluster0[i][j]==numpy.inf or numpy.isnan(cluster0[i][j]):
                cluster0[i][j]=max_value0
    min_sum0=cluster0.sum(axis=1).tolist()
    min_dists_per_feature[feature_name]['0']= dict_clusters[feature_name][0][min_sum0.index(min(cluster0.sum(axis=1)))]
    
    for i in range(len(cluster1)):
        for j in range(len(cluster1)):
            if cluster1[i][j]==numpy.inf or numpy.isnan(cluster1[i][j]):
                cluster1[i][j]=max_value1
    min_sum1=cluster1.sum(axis=1).tolist()
    min_dists_per_feature[feature_name]['1']= dict_clusters[feature_name][1][min_sum1.index(min(cluster1.sum(axis=1)))]
    

dict_small_big = {'adjusted_distances_from_lymphocytes_to_cancer_border':'small',
 'number_of_cells_in_lympho_blobs':'big',
 'lymphocyte_density':'small',
 'distance_between_lymphocytes_blobs':'big',
 'distances_from_lymphocytes_to_cancer':'small'}

for feature in all_feature_names:
    print("=" * 114)
    title = feature
    print(" " * int((114 - len(title)) / 2) + title)
    print("=" * 114)    
    if feature != 'lymphocyte_density':
        patient_id = min_dists_per_feature[feature_name]['1']
        plot_central_histogram(feature,patient_id,'1')
        patient_id = min_dists_per_feature[feature_name]['0']
        plot_central_histogram(feature,patient_id,'0')
    else :
        
        patient_id = min_dists_per_feature[feature]['1']
        pyplot.figure(figsize=(7.5,4.8))
        m=all_features[patient_id][feature].keys()
        n=all_features[patient_id][feature].values()
        my_xticks = ['[-40,-20]','[-20,0]','[0,20]','[20,40]']
        pyplot.xticks(list(m), my_xticks)
        pyplot.bar(list(m), n)
        pyplot.xlabel(feature+ ' S1 ($\mu$m)')
        pyplot.ylabel('Number of occurrences')
        pyplot.show()
        pyplot.savefig(feature+'_cluster1.png')        
        patient_id = min_dists_per_feature[feature]['0']
        m=all_features[patient_id][feature].keys()
        n=all_features[patient_id][feature].values()
        my_xticks = ['[-20,0]', '[0,20]', '[-80,-60]', '[-60,-40]','[-40,-20]','[20,40]','[40,60]']
        pyplot.figure(figsize=(7.5,4.8))
        pyplot.xticks(list(m), my_xticks)
        pyplot.bar(list(m), n)
        pyplot.xlabel(feature+ ' S2 ($\mu$m)')
        pyplot.ylabel('Number of occurrences')
        pyplot.show()
        pyplot.savefig(feature+'_cluster2.png')        
    try :
        compute_cluster_statistics(feature,dict_small_big[feature])
    except :
        pass

    
```

<h3>Cluster patients using distances from combination of multiple features</h3>

```python
dict_p_values = {}
dict_p_values['adjusted_distances_from_lymphocytes_to_cancer_border'] = 0.00021069325969752707
dict_p_values['number_of_cells_in_lympho_blobs'] = 0.015109622589428951
dict_p_values['lymphocyte_density'] = 0.00046079100681208106
dict_p_values['distance_between_lymphocytes_blobs'] = 0.01236070677669048

all_feature_names_voting = ['adjusted_distances_from_lymphocytes_to_cancer_border',
 'number_of_cells_in_lympho_blobs',
 'lymphocyte_density',
 'distance_between_lymphocytes_blobs']

voting_matrix = numpy.zeros((94,94))

## Apply log ponderation based on previously obtained p-value

for feature_name in all_feature_names_voting:
    for i in range(len(all_patients_ids)):
        for j in range(len(all_patients_ids)):
            patient_id_i = all_patients_ids[i]
            patient_id_j = all_patients_ids[j]
            if not ((patient_id_i in dict_clusters[feature_name][0] and  patient_id_j in dict_clusters[feature_name][0]) or (patient_id_i in dict_clusters[feature_name][1] and  patient_id_j in dict_clusters[feature_name][1])):
                voting_matrix[i][j]+= numpy.log(1/dict_p_values[feature_name])
numpy.fill_diagonal(voting_matrix, 0)          



# Using hierarchical clustering (merge clusters above threshold, find best threshold exhaustively)
# Header
print("=" * 114)
title = "Feature Combination"
print(" " * int((114 - len(title)) / 2) + title)
print("=" * 114)     

ratio_per_threshold = []
p_value_per_threshold = []
patient_clusters_per_threshold = []
nb_clusters = 1
while True :
    try :
        
        nb_clusters += 1
        patient_clusters = hierarchical_clustering(voting_matrix, all_patients_ids, nb_clusters=nb_clusters)
        patient_clusters.sort(key=len)
        patient_clusters = [flatten(patient_clusters[:-1]), patient_clusters[-1]]
        p_value = log_rank_test(patient_clusters, clinical_data)[1][1][0]
        ratio = len(patient_clusters[-1]) / (len(patient_clusters[-1]) + len(patient_clusters[0]))
        ratio_per_threshold.append(ratio)
        p_value_per_threshold.append(p_value)
        patient_clusters_per_threshold.append(patient_clusters)
    except :
        break
plot_curve(ratio_per_threshold, p_value_per_threshold, xlabel="Bottom cluster size", ylabel="Log-rank test", file_name=OUTPUT_DIR + "/clustering/" +"votingthresholds.png")
best_index = numpy.argmin(p_value_per_threshold)
plot_survival_curves(patient_clusters_per_threshold[best_index], clinical_data, file_name=OUTPUT_DIR + "/clustering/" +"votingsurvival.png")
p_values = log_rank_test(patient_clusters_per_threshold[best_index], clinical_data)
print('p-value :',p_values[1][1][0])
```
